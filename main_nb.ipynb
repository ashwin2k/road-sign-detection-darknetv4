```python
def create_txts():
    for index,row in train_annotation.iterrows():
        object_id=str(row['ClassId'])
        center_x="{:.5f}".format(int(row['Roi.X1']+row['Roi.X2'])/(2*48))#mean= divided by 2, normailizing between 0 and 1 , dividing by 48
        center_y="{:.5f}".format(int(row['Roi.Y1']+row['Roi.Y2'])/(2*56))
        width="{:.5f}".format(int(row['Width'])/48) 
        height="{:.5f}".format(int(row['Height'])/56)

        f = open("/home/ashwin/road-sign-detection/dataset/"+row['Path'][:-4]+".txt", "w")
        f.write(object_id+" "+center_x+" "+center_y+" "+width+" "+height)
        print("Created for %s"%row['Path'].split('/')[2])
        f.close()
```

```python
create_txts()
```

```python
def create_classes():
    meta=pd.read_csv("./dataset/Meta.csv")
    classes=meta['ClassId'].unique()
    f = open("./"+"classes"+".names", "w")
    for i in range(len(classes)):
        f.write(str(i)+"\n")
    f.close()
    print("Saved file at ./classes.names")
    print(len(classes))
```

```python
create_classes()
```

```python
def create_textfile_annotation():
    parent_dir='/home/ashwin/road-sign-detection/dataset/'
    train_paths=train_annotation['Path']
    p=[]
    for val in train_paths:
        path_to_save_into_txt_files=parent_dir+val
        p.append(path_to_save_into_txt_files + '\n')
    return p
```

```python
paths=create_textfile_annotation()
```

```python
data_test=paths[:int(len(paths) * 0.15)]
```

```python
data_train=paths[int(len(paths) * 0.15):]
```

```python
with open('train.txt', 'w') as train_txt:
    # Going through all elements of the list
    for e in data_train:
        # Writing current path at the end of the file
        train_txt.write(e)
```

```python
with open('test.txt', 'w') as test_txt:
    # Going through all elements of the list
    for e in data_test:
        # Writing current path at the end of the file
        test_txt.write(e)
```
